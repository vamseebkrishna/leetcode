```
2 Sum Problem: Given a sorted array of integers, find two numbers in the array that sum
to a given integer target.
For example, if a = [1,2,3,5,6,7] and target = 11, the answer will be 5 and 6.
```

```
public static (int, int)? TwoSum(int[] a, int target)
{
    if (a == null) return null;
    int start = 0;
    int end = a.Length-1;
    while (start < end)
    {
        int sum = a[start] + a[end];

        if (sum < target)
        {
            start++;
        }
        else if (sum > target)
        {
            end--;
        }
        else 
            return (a[start], a[end]);
    }
    return null;
}

//TEST
int[] a = [1,2,3,5,6,7];
int target = 11;
var result = TwoSum(a, target);
result
```
