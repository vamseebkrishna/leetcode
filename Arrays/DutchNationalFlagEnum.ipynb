```
Given an array with n marbles colored Red, White or Blue, sort them so that marbles of the same color are adjacent, with the colors in the order Red, White and Blue.
Assume the colors are given as numbers - 0 (Red), 1 (White) and 2 (Blue).
For example, if A = [1,0,1,2,1,0,1,2], Output = [0,0,1,1,1,1,2,2].
```

```
public enum Color {
    Red = 0,
    White = 1,
    Blue = 2
}

public class Marble 
{
    public Color Color { get; set; }
    public int Data { get; set; }

    public Marble(Color color, int data = 0)
    {
        this.Color = color;
        this.Data = data;
    }
    public override string ToString() => $"{Color}: {Data}";
}

public static Marble[] RedWhiteBlue(Marble[] a)
{
    if(a == null) return null;

    int low = 0;
    int high = a.Length-1;
    int i=0;

    while(i<=high)
    {
        switch(a[i].Color)
        {
            case Color.Red: Swap(a, i++, low++); 
                break;
            case Color.White: i++;
                break;
            case Color.Blue: Swap(a, i, high--);
                break;
            default: throw new ArgumentOutOfRangeException();
        }
    }
    return a;
}

public static void Swap(Marble[] a, int i, int j)
{
    Marble temp = a[i];
    a[i] = a[j];
    a[j] = temp;
}

//TEST
Marble[] collection = {
    new Marble(Color.Blue, 10),
    new Marble(Color.Red, 20),
    new Marble(Color.White, 30),
    new Marble(Color.Red, 40)
};

RedWhiteBlue(collection)
```
