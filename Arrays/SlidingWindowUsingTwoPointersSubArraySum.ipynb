```
Given an array of positive integers, find a subarray that sums to a given number X.
For e.g, input = [1,2,3,5,2] and X=8, Result = [3,5] (indexes 2,3)
```

```
public static (int, int)? SubArraySum(int[] a, int target) 
{
    if(a == null || a.Length == 0)
    {
        return null;
    }

    int start = 0, end = 0, sum = a[0];

    while(start < a.Length)
    {
        if(start > end)
        {
            end = start;
            if (start < a.Length)
            {
                sum = a[start];
            }
        }
        if (sum < target)
        {
            if (end == a.Length-1)
            {
                break;
            }
            end++;
            sum += a[end];
        }
        else if (sum > target)
        {
            sum = sum - a[start];
            start++;
        }
        else
        {
            return (start, end);
        }
    }
    return null;
}

//TEST
int[] arr = { 1, 2, 3, 7, 5 };
int target = 12;
var result = SubArraySum(arr, target);
result
```
