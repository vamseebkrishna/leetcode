```
You are given an array of integers. Rearrange the array so that all zeroes are at
the beginning of the array.
For example,
a = [4,2,0,1,0,3,0] -> [0,0,0,4,1,2,3]
```

```
public static int[] PartitionArrayBegining(int[] a)
{
    int boundry = 0;
    for (int i = 0; i < a.Length; i++)
    {
        if (a[i]==0)
        {
            Swap(a, i, boundry);
            boundry = boundry + 1;
        }
        
    }
    return a;
}

public static void Swap(int[] a, int i, int boundry)
{
    int temp = a[i];
    a[i] = a[boundry];
    a[boundry] = temp;
}

//TEST
int[] a = { 1, 0, 2, 0, 3, 0 };
PartitionArrayBegining(a);
a
```
