```
Given an array of integers that can be both +ve and -ve, find the contiguous subarray
with the largest sum.
For example: [1,2,-1,2,-3,2,-5] -> first 4 elements have the largest sum. Return (0,3)
```

```
public static (int start, int end) MaxSumSubArray(int[] a)
{
    if (a == null || a.Length == 0)
    {
        throw new ArgumentException("Input Array is empty or null");
    }

    int maxSum = a[0];
    int maxSumEndingHere = a[0];
    int start = 0;
    int end = 0;
    int tempStart = 0;

    for (int i = 0; i < a.Length; i++)
    {
        if(a[i]> maxSumEndingHere+a[i])
        {
            maxSumEndingHere = a[i];
            tempStart = i;
        }
        else
        {
            maxSumEndingHere = maxSumEndingHere + a[i];
        }
        if (maxSumEndingHere > maxSum)
        {
            maxSum = maxSumEndingHere;
            start = tempStart; 
            end = i;
        }
    }
    return (start, end);
}

//TEST
int[] input = { 1, 2, -1, 2, -3, 2, -5 };
var result = MaxSumSubArray(input);
result
```
