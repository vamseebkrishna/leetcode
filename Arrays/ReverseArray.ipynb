```
Given an array, reverse the order of its elements.
For example, [3,5,2,5,2,3,9] â†’ [9,3,2,5,2,5,3]
```

```
public static int[] ReverseArray(int[] a)
{
    if (a == null || a.Length == 0)
    {
        return a;
    }
    int start = 0;
    int end = a.Length-1;
    while(start < end)
    {
        Swap(a, start, end);
        start++;
        end--;
    }
    return a;
}

private static void Swap(int[] a, int i, int j)
{
    int temp = a[i];
    a[i] = a[j];
    a[j] = temp;
}

//TEST
int[] a = [1, 2, 3, 4, 5];
ReverseArray(a);
a
```
