```
Given an array of integers, rearrange the elements such that all zeros are
moved to the end of the array.
```

```
public static int[] PartitionArrayEnd(int[] a)
{
    if(a == null) return;

    int boundry = a.Length-1;

    for(int i=a.Length-1; i<=0; i--)
    {
        if(a[i] == 0)
        {
            Swap(a, i, boundry);
            boundry--;
        }
    }
    return a;
}

public static void Swap(int[] a, int i, int boundry)
{
    int temp;
    temp = a[i];
    a[i] = boundry;
    a[boundry] = temp;
}

//TEST
int[] a = [1, 0, 2, 0, 3, 0];
PartitionArrayEnd(a);
a
```
