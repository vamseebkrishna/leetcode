```
Dutch National Flag Problem: Given an array of integers A and a pivot, rearrange A in the following order:
[Elements less than pivot, elements equal to pivot, elements greater than pivot]

For example, if A = [5,2,4,4,6,4,4,3] and pivot = 4 -> result = [3,2,4,4,4,4,6,5]
```

```
public static int[] DutchNationalFlagPivot(int[] a, int pivot)
{
    int low = 0;
    int high = a.Length-1;
    int i = 0;
    while (i < high)
    {
        if(a[i]<pivot)
        {
            Swap(a, i, low);
            low++;
            i++;
        }
        else if(a[i]>pivot)
        {
            Swap(a, i, high);
            high--;
        }
        else
        {
            i++;
        }
    }
    return a;
}

public static void Swap(int[] a, int i, int j)
{
    int temp;
    temp = a[i];
    a[i] = a[j];
    a[j] = temp;
}

//TEST
int[] a = [5,2,4,4,6,4,4,3];
int pivot = 4;
DutchNationalFlagPivot(a, pivot);
a
```
