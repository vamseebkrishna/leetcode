```
Given an array of integers, both -ve and +ve, find a contiguous subarray that sums to a number
X.
For example: [2,4,-2,1,-3,5,-3], X = 5 --> Result = [2,4,-2,1]
```

```
public static (int start, int end)? TargetSumSubArray(int[] a, int target)
{
    if(a == null || a.Length ==0)
    {
        return null;
    }

    int sum = 0;
    var map = new Dictionary<int, int>();

    for(int i=0; i < a.Length; i++)
    {
        sum = sum + a[i];

        if (sum == target)
        {
            return(0,i);
        }

        int lookFor = sum-target;
        
        if(map.ContainsKey(lookFor))
        {
            return (map[lookFor] + 1, i);
        }
        if(!map.ContainsKey(sum))
        {
             map[sum] = i;   
        }
    }
    return null;
}

// TEST
int[] arr = { 10, 2, -2, -20, 10 };
int target = -10;
var result = TargetSumSubArray(arr, target);
result
```
